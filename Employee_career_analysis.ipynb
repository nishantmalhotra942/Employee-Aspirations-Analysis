```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the csv file
df = pd.read_csv('/content/Your Career Aspirations of GenZ.csv')
```

```python
print(df.columns)

print(df.info())
print(df.shape)
```

```python
'''AS WE ARE HAVING OBJECT DATATYPES WHICH WILL HAVE CATEGORIAL DATA SO WE ARE CHECKING A DIFFERENT TYPES OF VALUES THAT THEY CONTAIN'''
print(df.describe(include = 'object'))
for cols in df.describe(include = 'object').columns:
  print(cols)
  print(df[cols].unique())
  print('-'*50)
```

```python
print(df.isnull().sum())
print("NUMBER OF ROWS THAT ARE DUPLICATED",df.duplicated().sum())
```

```python
#Q1.) Would you definitely pursue a Higher Education / Post Graduation outside of India ? If only you have to self sponsor it.
high_education = df['Would you definitely pursue a Higher Education / Post Graduation outside of India ? If only you have to self sponsor it.'].value_counts()
mylabels = high_education.index
mycolors = ['red','green','blue']
myexplode= [0,0,0.2]
plt.title('HIgher EDUCATION')
plt.pie(high_education,labels=mylabels,colors=mycolors,autopct='%1.2f%%',explode=myexplode,shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9}, startangle=90)
```

```python
#Q2.)Which of the below factors influence the most about your career aspirations ?
print(df['Which of the below factors influence the most about your career aspirations ?'].value_counts())
```

```python
labels = ['My Parents ','People who have changed the world for better ' ,'People from my circle, but not family members ','Influencers who had successful careers ','Social Media like LinkedIn']
plt.pie((df['Which of the below factors influence the most about your career aspirations ?'].value_counts()),explode=[0,0.1,0.2,0.2,0],labels= labels,autopct='%1.2f%%',shadow={'ox': -0.04, 'edgecolor': 'black', 'shade': 0.9}, startangle=90)
```

```python
#ANALYSIS OF THE COUNTRY
curr_country = df['Your Current Country.'].value_counts()
print(curr_country)
labels = curr_country.index
mycolor = ['red','green','blue','yellow']
plt.xticks(rotation= 90)
plt.bar((labels),curr_country,color=mycolor)
plt.legend()
```

```python
#Q3.)How likely is that you will work for one employer for 3 years or more ?
one_company = df['How likely is that you will work for one employer for 3 years or more ?'].value_counts()
my_label = one_company.index
plt.pie(one_company,labels = my_label,autopct='%1.2f%%',explode=[0,0.1,0.1],shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9})
```

```python
#Q4.)What is the most preferred working environment for you??
working_condition = df['What is the most preferred working environment for you.'].value_counts()
my_label = working_condition.index
plt.pie(working_condition,labels = my_label,autopct='%1.2f%%',shadow={'ox': -0.04, 'edgecolor': 'none', 'shade': 0.9})
```
